```python
with open('corpus.txt', 'r') as f: 
    book = f.read()
    corpus = book.lower()
```

```python
def freq_dict(text):
    d = {}
    text = text.split()
    for word in text: 
        if word in d: 
            d[word] += 1
        else: 
            d[word] = 1
    return d

fd = freq_dict(corpus)

from operator import itemgetter
sorted_freq = sorted(fd.items(), key=itemgetter(1), reverse=True)
print(sorted_freq[:100])
```
